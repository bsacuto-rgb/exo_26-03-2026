# Automate Google Sheets with 💫

Remember the ETL you built for Bright Books? They like it soo much (yes they did 😉) that they want you to automate the process!

## 🔧 Your task

Using Github Actions, automate the already built ETL process. If you don't have it, feel free to clone that repo:

* https://github.com/JedhaBootcamp/automate-with-gsheet

<Note type="important">

The code contains environment variables, especially ones that are supposed to be pointing to a **database**. This time you won't be able to use Docker compose as your runner is not your local machine. 

Therefore, you will need to create a PostgreSQL database: Use AWS RDS for this exercise

</Note>

Also the data comes from this spreadsheet:

* [**Financial Data (Google Sheet)**](https://docs.google.com/spreadsheets/d/1npJZrNb30OgCpLTNEVwSQw1T6QqdQvdPbEDp8s3iNcg/edit?usp=sharing)


<Note type="important">

Copy this spreadsheet into your Google Drive (File → Make a copy). This way you’ll have **your own editable version**.

</Note>

## STEP I: CLONE / FORK THE REPO

* **Create your own repository based on the code your cloned** (A fork also should work!)

## STEP II: GET YOUR GCP JSON KEY

1. Go to [Google Cloud Console](https://console.cloud.google.com/)
2. Create a **new project** for this exercise  
   → Click on **“Open Project Picker”** (top bar next to the logo) → **New Project**
3. Inside that project, create a **Service Account**  
   → *IAM & Admin → Service Accounts → Create*  
   (no special roles or principals needed)
4. Generate a **Service Account JSON Key**  
   → On the right side of your service account row, click the 3 dots → **Manage Keys**  
   → **Add Key → Create new key → JSON**  
   → Download it and save it in your working directory.
5. **Enable the Google Sheets API**  
   → *APIs & Services → Enable APIs and Services → Search “Google Sheets API” → Enable*

## STEP III: CREATE AND SET YOUR DATABASE

* Then you need to set your database. Follow the AWS Tutorial 

## STEP IV: SET YOUR WORKFLOW

* Once you have that, you can create your workflow. 
* Run your workflow 
* And test it by connecting to your DB `docker run -it --rm postgres:16 psql -h <rds-endpoint> -U jedha -d <db-name>` 
* You should be prompted to your postgresql shell, run: `SELECT * FROM <t>able-name> LIMIT 5` (Replace `<table-name>` by the name of the table you chose in your environment variables)
